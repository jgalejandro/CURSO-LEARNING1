```python
import urllib
import numpy as np
import matplotlib.pyplot as plt
lienzo = plt.imread('/content/imagenes/AFALogo.bmp')
plt.imshow(lienzo, cmap = 'gray')
```

```python
#recorte de las letras
recorte_letras = lienzo[80:160,25:105]
plt.imshow(recorte_letras, cmap ='gray')
```

```python
#recorte y rotacion de estrellas
recorte_estrellas = lienzo[0:37,20:110]
algo = recorte_estrellas.T
plt.imshow(algo, cmap='gray')
```

```python
#espejado
plt.imshow(lienzo[:,::-1], cmap='gray')
```

```python
#mascara
lienzo2 = np.where(lienzo == 0, 0,255)
plt.imshow(lienzo2, cmap='gray')
```

```python
#fondo blanco
lienzo3 = np.where(lienzo == 0,255,lienzo)
plt.imshow(lienzo3, cmap='gray')
```

```python
#rejilla
dx = 2
dy = 5
lienzo4 = np.array(lienzo)
lienzo4[:,::dy] = 0
lienzo4[::dx,:] = 0
plt.imshow(lienzo4, cmap='gray')
```
